## Consultar la información

Para consultar la información debemos, en primer lugar, acceder a la función "request" de la librería.

Para comenzar consultaremos los reportes asociados a una entidad en particular, para esto debemos ingresar el rut de esta bajo el argumento "item" de la función. Como ejemplo, buscaremos los reportes asociados al Banco de Chile, por lo que bajo el argumento "item" pondremos el rut de la entidad ("97004000-5").

```python
data = pacifico.request(token, item='97004000-5')
data
```

Como podemos ver, existen una gran variedad de reportes asociados al Banco de Chile, estos están organizados en la jerarquia Document-Chapter-Section-Subsection-Paragraph y tienen o no valores según corresponda la clasificación de cada reporte.

En los items también podemos encontrar instrumentos, como por ejemplo buscaremos el bono "BCP0600323".

```python
data = pacifico.request(token, item='BCP0600323')
data
```

De esta forma obtenemos todo tipo de características sobre el instrumento, incluyendo la tabla de desarrollo, clasificaciones y fechas relevantes.

De la misma manera podemos buscar por otros filtros dentro de la jerarquía de los reportes, la categoría principal es el documento, donde podemos encontrar ciertas características para todas las entidades. Por ejemplo, podemos buscar todas las identificaciones, para esto debemos ingresar "Identification" bajo el argumento "document".

```python
data = pacifico.request(token, document='Identification')
data
```

También podríamos ver los calendarios disponibles en los reportes ingresando "Calendar" bajo el argumento "document".

```python
data = pacifico.request(token, document='Calendar')
data
```

Otro caso podria ser consultar los reportes sobre los balances sobre una empresa, por ejemplo Enel Distribucion Chile S.A. (96800570-7).

```python
data = pacifico.request(token, item='96800570-7', document='Balance Sheet')
data
```

## Metadata

Para saber cuál es el universo existente de datos para solicitar, debemos solicitar la metadata. Esta nos mostrara los autores, documentos e items disponibles en el mismo formato en el que recibimos datos anteriormente.

Para solicitar la metadata debemos usar la función "request" y pasar en alguno de los argumentos de los reports ("document", "chapter", "section", "subsection", "paragraph" o "item") el string "metadata".

```python
metadata = pacifico.request(token, document='metadata')
metadata
```

También podemos filtrar sobre la metadata, esto lo hacemos filtrando como de costumbre con los argumentos de la función "request", pero agregando el string "metadata" en alguno de los argumentos. Por ejemplo, consultaremos la metadata exclusivamente para el Banco de Chile.

```python
metadata = pacifico.request(token, document='metadata', item='97004000-5')
metadata
```

De esta forma sabemos que tipos de reportes están registrados para el Banco de Chile.

## Autores

Como se puede observar en la sección de la metadata, existen varios autores de datos dentro de la API, por lo que podemos especificar el autor o fuente de los datos al solicitarlos. Para hacer esto debemos especificar el autor bajo el argumento "author".

Como ejemplo pediremos la metadata de los reportes del Banco Central de Chile (BCCh).

```python
data_bcch = pacifico.request(token, author='bcch', document='metadata')
data_bcch
```

## Convertir a otros formatos

Por defecto las consultas a la API mediante la librería devuelven un Pandas DataFrame con la información consultada. Pero existen otros dos formatos en que la API puede devolver esta información, en un diccionario de Python o en un string en formato JSON.

La manera en que especificamos el formato de salida de datos es mediante el argumento "format". Para el cual podemos especificar los 3 formatos como "dataFrame", "dictionary" y "json".

Ejecutaremos la misma consulta inicial para el bono 'BCP0600323' en los 3 distintos formatos. 

Para el Pandas DataFrame (por defecto):

```python
data = pacifico.request(token, item='BCP0600323', format='dataFrame')
data
```

Para un diccionario de Python:

```python
data = pacifico.request(token, item='BCP0600323', format='dictionary')
print(data)
```

Para el string en formato JSON:

```python
data = pacifico.request(token, item='BCP0600323', format='json')
print(data)
```

Así existe flexibilidad en la entrega de datos por parte de la librería.

## Guardar

Finalmente existen dos formas en las cuales podemos guardar los datos pedidos en una consulta, estos son como un archivo de texto o un archivo CSV y cual sea depende del formato en que se pidieron los datos.

Para poder guardar los datos debemos especificar bajo el argumento "fileName" el nombre del archivo donde los almacenaremos.

Para que la consulta se grabe como archivo de texto debemos pedir los datos en formato JSON o diccionario de Python y se guardaran con estructura JSON. Por otro lado, para que la consulta se grave como un archivo CSV debemos pedir los datos en formato Pandas DataFrame.

Volveremos a realizar el ejemplo para el bono 'BCP0600323' y lo guardaremos de las dos formas.

Para un archivo de texto:

```python
data = pacifico.request(token, item='BCP0600323', format='json', fileName='pacifico_data')
```

Para un archivo CSV:

```python
data = pacifico.request(token, item='BCP0600323', format='dataFrame', fileName='pacifico_data')
```

Por defecto los archivos serán guardados donde se está ejecutando este notebook o la librería.

En este caso luego de ejecutar los comandos tendremos los archivos "pacifico_data.txt" y "pacifico_data.csv", los cuales contienen los mismos datos, pero en distintos formatos.

## Descripción de los datos

### Agrupaciones de datos

Los datos que entrega la API tienen una jerarquía definida por el formato del archivo de entrega. Esto nos permite administrar los datos ordenadamente. Esta jerarquía es la siguiente:

- Documento
    - Capitulo
        - Sección
            - Subsección
                - Parrafo
                    - Item

Así, cada Item pertenece a un Parrafo, cada Parrafo pertenece a una Subsección, cada Subsección pertenece a una Sección, cada Sección pertenece a un Capitulo y cada Capitulo pertenece a un Documento.

### Fechas

Existen distintos tipos de fechas que definen características cruciales de los datos. A continuación, una breve descripción de cada uno de ellos:

- **Fecha de Publicación y Fixing**: Define la fecha en que el dato fue publicado y el fixing de la publicación. Si la hora de fixing es distinta al fin del día de publicación, entonces la función devolvera la fecha de publicación y la hora del fixing en intervalos de media hora, de lo contrario devolvera sólo la fecha de publicación.


- **Fecha Efectiva**: Corresponde a la fecha efectiva para el valor de este dato, por lo general esta será igual a la fecha de publicación, pero no necesariamente. Por ejemplo, cuando un precio es públicado al día desdepués de ser transado o calculado.


- **Fecha de Tenor**: Esta fecha, para los casos en que corresponda, mostrará la fecha del tenor del dato en cuestión. Un buen ejemplo, es el caso de un factor de descuento, aquí es vital saber para qué fecha del tenor está calculado el factor de descuento, y por lo tanto se incluirá esta fecha.


- **Fecha de Streaming**: Esta fecha corresponde a la fecha y hora en que el dato fue consultado por el cliente.

## Formato de Entrega

#### JSON

De manera general, los datos consultados desde la API serán entregados en formato JSON, y en una estructura que permite ordenar los datos obtenidos en base a sus características. Esta estructura, almacena características que pueden compartir los datos en capas. Así, los datos quedan seccionados en categorías, haciendo el archivo de entrega más liviano y ordenado.

El orden de las características o capas y la estructura general de entrega de la API es la siguiente:

<body>
<code>{Autor: 
    {Fecha de publicación: 
        {Fecha efectiva: 
            {Documento: 
                {Capitulo: 
                    {Sección: 
                        {Subsección: 
                            {Parrafo:
                                {Item:
                                    [{
                                        "variant": {
                                                    "value": Valor del dato,
                                                    "type": Tipo de Dato
                                                    }, 
                                        "dateTenor": Fecha del tenor del dato (en el caso que corresponda),
                                        "other": Característica extra del dato (en el caso que corresponda)
                                    }]
                                }
                            }
                        }
                    }
                }
            }
        }
    }
}
</code>

## Argumentos

El proceso de consulta de datos tiene múltiples variables de especificación, a partir de las cuales podemos acotar nuestras consultas de manera eficiente, estos son precisamente los argumentos que le entregaremos a la API. A continuación, están estos argumentos y sus descripciones:

- **Document**: Corresponde al documento sobre los cuales se quieren consultar reportes (ej. 'Identification' para los reportes de identificación).


- **Chapter**: Corresponde al capítulo sobre los cuales se quieren consultar datos.


- **Section**: Corresponde a la sección sobre los cuales se quieren consultar datos.


- **Subsection**: Corresponde a la subsección sobre los cuales se quieren consultar datos.


- **Paragraph**: Corresponde al parrafo sobre los cuales se quieren consultar datos.


- **Item**: Corresponde al item o cosa sobre los cuales se quieren consultar datos (ej. '97004000-5' para el Banco de Chile o 'BCP0600323' para las características de este bono).


- **Fecha de Inicio**: Corresponde a la fecha de inicio cuando se quieren consultar datos históricamente.


- **Fecha de Término**: Corresponde a la fecha de término cuando se quieren consultar datos históricamente.


- **Fixing**: Corresponde al fixing o fixings sobre los cuales se quieren consultar datos (ej. 'F_0000' para las 00:00 horas).


- **Autor**: Corresponde al autor de los datos que se quieren consultar (ej. 'Pacifico' para cálculos y publicaciones hechas por Pacifico).


- **Versión**: Corresponde a la versión de los datos que se quieren consultar (ej. '' para la versión por defecto establecida por Pacifico).
