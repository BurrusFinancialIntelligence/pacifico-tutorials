### Usando un URL

Para solicitar información a la API mediante un URL en este se deben especificar los argumentos de la solicitud como parámetros de cadenas de consulta a la dirección **https://api.pacificoindices.com**. El único argumento obligatorio para solicitar información es el token, que debiese ya haber sido proporcionado en el archivo **token.key**. Este debiese verse así:

<img src="Documents/tokenfolder.png" width="80%" align="center"/>

Y al abrir el archivo **token.key** debiese verse así (token de ejemplo):
<img src="Documents/token.key.png" width="80%" align="center"/>

Este token de texto deben agregarlo bajo la llave "token" de la siguiente manera:

https://api.pacificoindices.com/?token=MiToken

Al hacer clic en el URL automáticamente se hará un GET request a este y devolverá un enlace en donde se podrán descargar los datos consultados. En estos ejemplos el token es solo para mostrar cómo se agrega al URL y no es válido, por lo que al hacerle clic o un GET request este devolverá **{"message":"Forbidden"}**.

De esta manera podemos consultar distintos datos en base a los argumentos que agreguemos al URL. Por ejemplo, para consultar el valor de un instrumento especifico debemos ingresar el nombre de este bajo el argumento "ticker" de la función. Como ejemplo, buscaremos el bono 'BCP0600221' que corresponde a un bono del Banco Central de Chile ('BCP').

https://api.pacificoindices.com/?token=MiToken&ticker=BCP0600221

También podemos buscar datos de la familia a la que pertenece este bono.

https://api.pacificoindices.com/?token=MiToken&family=BCP

También podemos buscar datos del grupo de esta familia y bono.

https://api.pacificoindices.com/?token=MiToken&group=BT

También podemos buscar datos del mercado de este grupo, familia y bono.

https://api.pacificoindices.com/?token=MiToken&market=RF

### Usando la librería de Python

Particularmente para la librería de Python existen dos formas de uso, a través del mismo Python o mediante el command line.

### Mediante Python

##### 1) Importar la librería Pacifico Índices

```python
import pacifico
```

Esta librería incluye todas las funcionalidades disponibles de la API en forma de funciones de Python.

##### 2) Autentificarse con el token

La librería, por defecto buscará el token con el fin de realizar la autentificación. Este se encuentra en un archivo plano de texto con nombre **token.key**, ubicado en la misma carpeta desde donde se está ejecutando la librería. Además, es posible especificar el token o el nombre del archivo que contiene el token directamente en las funciones de la librería bajo el argumento "token".

##### 3) Consultar la información

Para consultar la información debemos, en primer lugar, acceder a la función "request" de la librería.

Si la función se ejecuta sin ningún argumento, esta entregará por defecto los últimos precios publicados.

Para consultar el valor de un instrumento especifico debemos ingresar el nombre de este bajo el argumento "ticker" de la función. Como ejemplo, buscaremos el bono 'BCP0600221' que corresponde a un bono del Banco Central de Chile ('BCP').

```python
data = pacifico.request(ticker='BCP0600221')
data
```

También podemos buscar datos de la familia a la que pertenece este bono.

```python
data_family = pacifico.request(family='BCP')
data_family
```

También podemos buscar datos del grupo de esta familia y bono.

```python
data_group = pacifico.request(group='BT')
data_group
```

Tambien podemos buscar datos del mercado de este grupo, familia y bono.

```python
data_market = pacifico.request(market='RF')
data_market
```

Al ejecutar la función para obtener datos, esta nos devuelve por defecto un Pandas Dataframe con la información consultada.

#### 4) Convertir a otros formatos

Por defecto las consultas a la API mediante la librería devuelven un Pandas DataFrame con la información consultada. Pero existen otros dos formatos en que la API puede devolver esta información, en un diccionario de Python o en un string en formato JSON.

La manera en que especificamos el formato de salida de datos es mediante el argumento "format". Para el cual podemos especificar los 3 formatos como "dataFrame", "dictionary" y "json".

Ejecutaremos la misma consulta inicial para el bono 'BCP0600221' en los 3 distintos formatos. 

Para el Pandas DataFrame (por defecto):

```python
data = pacifico.request(ticker='BCP0600221', format='dataFrame')
data
```

Para un diccionario de Python:

```python
data = pacifico.request(ticker='BCP0600221', format='dictionary')
print(data)
```

Para el string en formato JSON:

```python
data = pacifico.request(ticker='BCP0600221', format='json')
print(data)
```

Así existe flexibilidad en la entrega de datos por parte de la librería. 

#### 5) Guardar

Finalmente existen dos formas en las cuales podemos guardar los datos pedidos en una consulta, estos son como un archivo de texto o un archivo CSV y cual sea depende del formato en que se pidieron los datos.

Para poder guardar los datos debemos especificar bajo el argumento "fileName" el nombre del archivo donde los almacenaremos.

Para que la consulta se grabe como archivo de texto debemos pedir los datos en formato JSON o diccionario de Python y se guardaran con estructura JSON. Por otro lado, para que la consulta se grave como un archivo CSV debemos pedir los datos en formato Pandas DataFrame.

Volveremos a realizar el ejemplo para el bono 'BCP0600221' y lo guardaremos de las dos formas.

Para un archivo de texto:

```python
data = pacifico.request(ticker='BCP0600221', format='json', fileName='pacifico_data')
```

Para un archivo CSV:

```python
data = pacifico.request(ticker='BCP0600221', format='dataFrame', fileName='pacifico_data')
```

Por defecto los archivos serán guardados donde se está ejecutando este notebook o la librería.

En este caso luego de ejecutar los comandos tendremos los archivos "pacifico_data.txt" y "pacifico_data.csv", los cuales contienen los mismos datos, pero en distintos formatos.

## Consulta Historica

Para consultar datos históricos, la API sólo tomará en cuenta las consultas que especifiquen fechas si es que también especifican los tickers a consultar. En el caso que no sea así devolverá los datos consultados, pero de hoy o la última fecha disponible, ignorando la especificación histórica.

### Usando un URL

En el caso de un URL, para hacer una consulta historica debemos especificar la fecha de inicio y termino de la consulta mediante los argumentos "dateStart" y "dateEnd" respectivamente.

Como ejemplo, el URL para la acción del Banco de Chile históricamente desde el 16 de septiembre hasta el 16 de octubre del 2020 sera el siguiente:

https://api.pacificoindices.com/?token=MiToken=CHILE&dateStart=16/9/2020&dateEnd=16/10/2020

### Usando la librería mediante Python

En el caso de la librería de Python, hay que introducir las fechas como un objeto "datetime" en la función "request".

Como ejemplo, buscaremos el precio de la acción del Banco de Chile para los últimos 7 días.

```python
import datetime

fechaInicio = datetime.date.today() - datetime.timedelta(days=7)
fechaTermino = datetime.date.today()

data = pacifico.request(ticker='CHILE', dateStart=fechaInicio, dateEnd=fechaTermino)
data
```

### Usando la librería mediante command line

En el caso del command line, hay que introducir las fechas como argumentos con un '-' antes del nombre de los argumentos, en el caso de la fecha de inicio el argumento se llama "dateStart" y en el caso de la fecha de término "dateEnd".

Como ejemplo, buscaremos la misma acción del Banco de Chile, pero históricamente desde el 16 de septiembre hasta el 16 de octubre del 2020.

```python
!python pacifico.py -ticker CHILE -dateStart 16/9/2020 -dateEnd 16/10/2020
```

### Fechas

Existen distintos tipos de fechas que definen características cruciales de los datos. A continuación, una breve descripción de cada uno de ellos:

- **Fecha de Publicación y Fixing**: Define la fecha en que el dato fue publicado y el fixing de la publicación. Si la hora de fixing es distinta al fin del día de publicación, entonces la función devolvera la fecha de publicación y la hora del fixing en intervalos de media hora, de lo contrario devolvera sólo la fecha de publicación.


- **Fecha Efectiva**: Corresponde a la fecha efectiva para el valor de este dato, por lo general esta será igual a la fecha de publicación, pero no necesariamente. Por ejemplo, cuando un precio es públicado al día desdepués de ser transado o calculado.


- **Fecha de Tenor**: Esta fecha, para los casos en que corresponda, mostrará la fecha del tenor del dato en cuestión. Un buen ejemplo, es el caso de un factor de descuento, aquí es vital saber para qué fecha del tenor está calculado el factor de descuento, y por lo tanto se incluirá esta fecha.


- **Fecha de Streaming**: Esta fecha corresponde a la fecha y hora en que el dato fue consultado por el cliente.

### Formato de Entrega

#### JSON

De manera general, los datos consultados desde la API serán entregados en formato JSON, y en una estructura que permite ordenar los datos obtenidos en base a sus características. Esta estructura, almacena características que pueden compartir los datos en capas. Así, los datos quedan seccionados en categorías, haciendo el archivo de entrega más liviano y ordenado.

El orden de las características o capas y la estructura general de entrega de la API es la siguiente:

<body>
<code>{Escenario: 
    {Fecha de publicación: 
        {Fecha efectiva: 
            {País: 
                {Mercado: 
                    {Grupo: 
                        {Familia: 
                            {Ticker:
                                [{
                                    "value": Valor del dato, 
                                    "field": Tipo de Dato,
                                    "dateTenor": Fecha del tenor del dato (en el caso que corresponda),
                                    "other": Característica extra del dato (en el caso que corresponda)
                                }]
                            }
                        }
                    }
                }
            }
        }
    }
}
</code>
</body>

Para poder ver esta estructura ejemplificada, tomaremos el bono 'BCP0600221' que obtuvimos en la primera sección del tutorial:

<body>
<code>{"Pricing": 
    {"05/10/2020": 
        {"05/10/2020 19:15:00": 
            {"Chile": 
                {"RF": 
                    {"BT": 
                        {"BCP": 
                            {"BCP0600221": 
                                [{
                                    "value": 102.92331175431464, 
                                    "field": "Price"
                                }, 
                                {
                                    "value": 0.0023, 
                                    "field": "Yield"
                                }]
                            }
                        }
                    }
                }
            }
        }
    }
}
</code>
</body>

Aquí podemos ver que el valor obtenido cuenta con un escenario de pricing, lo que significa que corresponde a un precio calculado (también está la posibilidad de que corresponda a una predicción). En este caso, no se incluye el nombre del autor del escenario debido a que corresponde a BFI. También, el resultado tiene como fecha de publicación el 5 de Octubre del 2020; la fecha efectiva del precio también corresponde al 5 de Octubre del 2020 y además incluye la hora 19:15:00; el país corresponde a Chile; el mercado corresponde a la renta fija (RF), el grupo corresponde a los Bonos Tesorería (BT), la familia corresponde a los Bonos de Banco Central de Chile (BCP); el ticker corresponde al mnemotécnico del bono; finalmente en la lista de datos tenemos dos datos, en primer lugar el valor (bajo la llave "value") que corresponde a 102.92331175431464 y el tipo de dato (bajo la llave "field") que corresponde a un precio, y en segundo lugar el valor (bajo la llave "value") que corresponde a 0.0023 y el tipo de dato (bajo la llave "field") que corresponde a una tasa.

### Argumentos

El proceso de consulta de datos tiene múltiples variables de especificación, a partir de las cuales podemos acotar nuestras consultas de manera eficiente, estos son precisamente los argumentos que le entregaremos a la API. A continuación, están estos argumentos y sus descripciones:

- **Ticker**: Corresponde al ticker o tickers sobre los cuales se quieren consultar datos (ej. 'CHILE' para la acción del Banco de Chile).


- **Familia**: Corresponde a la familia o familias sobre los cuales se quieren consultar datos (ej. 'BCP' para los Bonos de Banco Central de Chile).


- **Grupo**: Corresponde al grupo o grupos sobre los cuales se quieren consultar datos (ej. 'BT' para los Bonos de Tesorería).


- **Mercado**: Corresponde al marcado o mercados sobre los cuales se quieren consultar datos (ej. 'RF' para el mercado de Renta Fija).


- **País**: Corresponde al país o países sobre los cuales se quieren consultar datos (ej. 'Chile' para Chile).


- **Fecha de Inicio**: Corresponde a la fecha de inicio cuando se quieren consultar datos históricamente.


- **Fecha de Término**: Corresponde a la fecha de término cuando se quieren consultar datos históricamente.


- **Fixing**: Corresponde al fixing o fixings sobre los cuales se quieren consultar datos (ej. 'F_0000' para las 00:00 horas).


- **Tipo de Datos**: Corresponde al tipo de datos sobre los cuales se quieren consultar datos (ej. 'Price' para los precios).


- **Tipo de Versión**: Corresponde al tipo de versión de los datos que se quieren consultar (ej. 'Pricing' cálculos sobre precios).


- **Autor**: Corresponde al autor de los datos que se quieren consultar (ej. 'BFI' para cálculos y publicaciones hechas por Burrus Financial Intelligence).


- **Versión**: Corresponde a la versión de los datos que se quieren consultar (ej. '' para la versión por defecto establecida por Burrus Financial Intelligence).

Existen mercados especiales dentro del universo de los mercados, los que contienen todos los tickers que almacenan algún tipo de factor. Estos se consideran en un nivel mas bajo de datos ya que gracias a estos otros valores son calculados.
